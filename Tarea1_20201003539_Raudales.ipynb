# **Tarea DATAFRAME**

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
asignaturas =  ['POO', 'Algoritmos', 'BD', 'Teoria de la simulacion', 'compiladores', 'Programacion 1', 'Programacion 2',
                              'Redes de datos']
```

## Creacion de los Registros y poblacion del DATAFRAME 

```python
calificaciones = []
materias = []
for i in range (3000):
    materias.append(random.choice(asignaturas))
    calificaciones.append(random.randint(0,101))
df = pd.DataFrame({
    'Asignaturas': materias,
    'Calificacione': calificaciones
})

print(df)
```

## Resumen estadistico que muestra el numero de registros, la variacion y el valor minimo

```python
df.describe()
```

## Resumen Estadistico de cuantas veces se repite una clase con el value_counts

```python
resumenAsignaruras = df['Asignaturas'].value_counts()
print("Resumen de Asignaturas\n")
print(resumenAsignaruras) 
```

## Resumen Estadistico de variables mediante el include=['O'] que muestra valores no nulos, valores nulos y el valor que mas aparece

```python
r = df.describe(include=['O'])
print (r)
```

## Grafico de barras creado con **barplot** que muestra el promedio de calificaciones por materia

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Asignaturas', y='Calificacione', data=df, estimator=np.mean)
plt.title('Promedio de calificaciones')
plt.xlabel('Asignaturas')
plt.ylabel('Calificaciones')
plt.xticks(rotation=45)
plt.show()
```

## Grafico de cajas creado con **boxplot** muestra la distribucion de calificaciones por materias

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Asignaturas', y='Calificacione', data=df)
plt.title('Promedio de calificaciones')
plt.xlabel('Asignaturas')
plt.ylabel('Calificaciones')
plt.xticks(rotation=50)
plt.show()
```
